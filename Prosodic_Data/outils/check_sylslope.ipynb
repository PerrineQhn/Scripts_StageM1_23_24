```python
import os
```

```python
def check_syl_slope(file_path):
    valid_slopes = {"Rise", "Fall", "Flat", "X"}
    sentences_to_review = []

    with open(file_path, 'r', encoding='utf-8') as file:
        current_sentence_id = None
        for line in file:
            line = line.strip()
            if line.startswith("# sent_id"):
                current_sentence_id = line.split(" = ")[1]
            elif line.startswith("#"):
                continue
            elif line:
                columns = line.split("\t")
                if len(columns) > 9:  # Check if line has annotations
                    # print(columns[9])
                    for i in range(1, 9):  # Check up to 8 syllables
                        slope_key = f"Syl{i}Slope"
                        if any(slope_key in column and column.split("=")[1] not in valid_slopes for column in columns[9].split("|")):
                            sentences_to_review.append(current_sentence_id)
                            break  # No need to check further syllables in this word
            else:
                current_sentence_id = None  # Reset for the next sentence

    return set(sentences_to_review)
```

```python
# directory_conllu = "./conllu_syllables_30janv/"
directory_conllu = "./SUD_Naija-NSC-master"

for file in os.listdir(directory_conllu):
    if file.endswith(".conllu"):
        file_path = os.path.join(directory_conllu, file)
        sentences_to_review = check_syl_slope(file_path)
        print(f"File: {file}")
        if sentences_to_review:
            print(f"Sentences to review in {file}:")
            for sentence in sentences_to_review:
                print(sentence)
            print()
```
